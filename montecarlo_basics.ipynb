```python
# Importing useful pakages
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import sympy as smp
plt.style.use(['science','notebook'])
```

```python
# now let's define the given PDF and CDF
x = np.linspace(0,7,100)
f = 3*np.exp(-3*x)
F = 1-np.exp(-3*x)
```

```python
f
```

```python
F
```

```python
plt.figure(figsize=(15,8))
plt.plot(x, f, label=r'$f(x)$')
plt.plot(x,F, label=r'$F(x)$')
plt.legend()
plt.xlabel('$x$', fontsize=30)
plt.show()
```

```python
# U - uniform variable
U = np.random.rand(10000) # generates random numbers from the standard uniform distribution
#(i.e., the uniform distribution from 0 to 1)
U
```

```python
plt.hist(U) # this is the generated distribution of random variable U
```

```python
# now we will define F_inverse , lambda = 3
F_inv_U = (-np.log(1-U))/3
```

```python
F_inv_U
```

```python
plt.hist(F_inv_U) 
plt.ylabel("$U$")
plt.xlabel("$F^{-1}(u)$")
```

```python
# let's see the distribution of f(x) and F_inv(x)
plt.figure(figsize=(15,8))
plt.plot(x, f, label=r'$f(x)$',color='red')
plt.hist(F_inv_U, histtype='step', color='green',density='norm', bins=100, label='$F^{-1}(u)$')
plt.legend()
plt.xlabel('$x$', fontsize=20)
plt.show()
# Here, we got the distribution as we desired it to be.
```
